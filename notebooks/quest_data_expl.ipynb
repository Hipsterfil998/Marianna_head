```python
import pandas as pd
import ast
```

```python
def open_file(file_path):

    df = pd.read_csv(file_path, delimiter='\t')
    return df

a = open_file('/content/drive/MyDrive/testa_di_Marianna/dati_wiki_ristretti/wiki_naples_QA.tsv')
```

```python
a
```

```python
questions = a['question'].tolist()
```

```python
questions = [list(filter(None, ast.literal_eval(el))) for el in questions]
```

```python
questions
```

```python
c = 0

for el in questions:
  for e in el:
    c += 1

c
```

```python
questions_cleaned = []

for el in questions:
  question_list = el.strip("[]").replace("\\n", "\n").split("\n")
  question_list = [el.split(',') for el in question_list]
  questions_cleaned.append(question_list)
```

```python
questions_cleaned_2 = []

for g in questions_cleaned:
  list_q = []
  for el in g:
    for e in el:
      ques = e.strip(" ").strip('"').strip("'")
      list_q.append(ques)
  questions_cleaned_2.append(list_q)
```

```python
questions_cleaned_2[:3]
```

```python
dataq = {'titles': a['titles'].tolist(),
        'text': a['text'].tolist(),
        'question': questions_cleaned_2}


dfq = pd.DataFrame(dataq)
dfq.to_csv('/content/drive/MyDrive/testa_di_Marianna/dati_wiki_ristretti/wiki_naples_QA.tsv', sep='\t', index=False)
```
